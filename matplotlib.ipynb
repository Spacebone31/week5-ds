```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
data = load_iris(as_frame=True)
iris = data.data
target = data.target
```

OUTPUT NO.1 SCATTER PLOT

```python
plt.figure(figsize=(6, 4))
plt.title('Scatter Plot of sepal_length and petal_length')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.scatter(iris['sepal length (cm)'], iris['petal length (cm)'], c=target)
```

OUTPUT NO.2 HISTOGRAM

```python
iris['target'] = target
iris['species'] = iris['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python
# Plot stacked histogram using seaborn
sns.histplot(
    data=iris, 
    x='sepal length (cm)',
    hue='species', 
    multiple='stack',
    stat='density',
    bins=[4.5, 5.0, 5.5, 6.0, 6.5, 7.0, 7.5, 8.0],
)

# Label and title
plt.ylabel('Density')
plt.title('Histogram of sepal_length')

# Show the plot
plt.show()
```

OUTPUT NO.3 PLOT BOX

```python
plt.figure(figsize=(6, 4))
plt.title('Box Plot of sepal_length')

box = plt.boxplot([iris[iris['species'] == 'setosa']['sepal length (cm)'],
             iris[iris['species'] == 'virginica']['sepal length (cm)'],
             iris[iris['species'] == 'versicolor']['sepal length (cm)']],
            labels=['Setosa', 'Versicolor', 'Virginica'],
            patch_artist=True,
            widths=0.5)

colors = ['green', 'purple', 'yellow']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```
